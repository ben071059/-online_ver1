# 人生online模擬器ver_1，只考慮有無投資的差異

```python
人生第一桶金 = 100 #單位萬
每月薪水 = 4.5
每月花費 = 2 #含孝親費+娛樂+必須
退休年齡 = 65
時段 = range(30,90,1)
```

```python
import matplotlib.pylab as plt
import pandas as pd
#每年淨額
每年淨額 = pd.Series(0,index = 時段 )
每年淨額.iloc[0] = 人生第一桶金
每年淨額.loc[:退休年齡] += 每月薪水*12
每年淨額 -= 每月花費*12
%matplotlib inline
每年淨額.plot()
```

```python
#計算如果不投資的資產變化
無投資財產 = 每年淨額.cumsum()
無投資財產.plot()
```

若有投資會發生甚麼事情呢?

簡單說明想法:
投資金額 = 淨額 乘上 投資部位
其餘部位(不動的錢) = 淨額 乘上 (1-投資部位)

```python
投資部位 = 0.5
投資年利率 = 5 #單位(%)

def invest(money,ratio,interest_rate):
    profit = [money.iloc[0]]
    for i in money[1:]:
        profit.append(profit[-1]*ratio*(interest_rate/100 + 1) + profit[-1]*(1-ratio) + i)
    return pd.Series(profit,index = 時段)
有投資財產 = invest(每年淨額,投資部位,投資年利率)
有投資財產.plot()
```

比較有投資跟無投資的財富曲線

```python
pd.DataFrame({"if_invest":有投資財產,
             "no_invest" : 無投資財產}).plot()

#有投資財產.plot()
#無投資財產.plot()
```

```python
薪資不可能一成不變，若加入調薪機制
```

```python
調整薪資 = pd.Series(0,index = 時段 )
薪資成長率 = 3 #單位(%)
薪資停止成長年齡 = 40
調整薪資.iloc[0] = 每月薪水*12
for i in range(30+1,退休年齡):
    調整薪資.loc[i] = 調整薪資.loc[i-1]*(薪資成長率/100 + 1) if i < 薪資停止成長年齡 else 調整薪資.loc[i-1]
```

```python
#每年淨額
每年淨額 = pd.Series(0,index = 時段 )
每年淨額.iloc[0] = 人生第一桶金
每年淨額.loc[:退休年齡] += 調整薪資
每年淨額 -= 每月花費*12
%matplotlib inline
每年淨額.plot()
```

```python
#計算如果不投資的資產變化
無投資財產 = 每年淨額.cumsum()
#無投資財產.plot()
#計算投資的資產變化
投資部位 = 0.5
投資年利率 = 5 #單位(%)
有投資財產 = invest(每年淨額,投資部位,投資年利率)
#有投資財產.plot()
pd.DataFrame({"if_invest":有投資財產,
             "no_invest" : 無投資財產}).plot()
```

了解未來的財富曲線

```python
import pandas as pd
def assets(人生第一桶金,
           開始年齡,
           退休年齡,
           初始每月薪水,
           薪資年調整率,
           薪資停止調整年齡,
           每月花費,
           投資部位,
           投資年利率):
    時段 = range(開始年齡,100)
    #薪資
    調整薪資 = pd.Series(0,index = 時段 )
    調整薪資.iloc[0] = 初始每月薪水*12
    for i in range(開始年齡+1,退休年齡):
        調整薪資.loc[i] = 調整薪資.loc[i-1]*(薪資年調整率/100 + 1) if i < 薪資停止調整年齡 else 調整薪資.loc[i-1]
    #每年淨額
    每年淨額 = pd.Series(0,index = 時段 )
    每年淨額.iloc[0] = 人生第一桶金
    每年淨額.loc[:退休年齡] += 調整薪資
    每年淨額 -= 每月花費*12
    #計算如果不投資的資產變化
    無投資財產 = 每年淨額.cumsum()
    def invest(money,ratio,interest_rate):
        profit = [money.iloc[0]]
        for i in money[1:]:
            profit.append(profit[-1]*ratio*(interest_rate/100 + 1) + profit[-1]*(1-ratio) + i)
        return pd.Series(profit,index = 時段)
    有投資財產 = invest(每年淨額,投資部位,投資年利率)
    pd.DataFrame({"if_invest":有投資財產,
             "no_invest" : 無投資財產}).plot()
        
import ipywidgets as widgets
widgets.interact(assets, 
    人生第一桶金=widgets.FloatSlider(min=0, max=200, step=10, value=20),
    開始年齡=widgets.IntSlider(min=0, max=100, step=1, value=30),
    初始每月薪水=widgets.FloatSlider(min=0, max=20, step=0.1, value=3),
    每月花費=widgets.FloatSlider(min=0, max=20, step=0.2, value=1),
    薪資年調整率=widgets.FloatSlider(min=0, max=20, step=0.1, value=1),
    薪資停止調整年齡=widgets.IntSlider(min=20, max=100, step=1, value=50),             
    退休年齡=widgets.IntSlider(min=0, max=100, step=1, value=60),
    投資部位=widgets.FloatSlider(min=0, max=1, step=0.1, value=0.7),
    投資年利率=widgets.FloatSlider(min=0, max=20, step=0.5, value=5),
    
)
           
```
